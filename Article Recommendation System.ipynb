```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df.duplicated().sum()
```

```python
df['Article_id'] = df.reset_index().index
df
```

```python
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
sw=set(stopwords.words('english'))
ps=PorterStemmer()
def clean_text(sample):
    sample=re.sub('[^A-Za-z0-9 ]',' ',str(sample))
    sample=sample.lower()
    sample=sample.split()
    sample=[ps.stem(s) for s in sample if s not in sw]
    sample=" ".join(sample)
    return sample
```

```python
df['Clean_Article']=df['Article'].apply(clean_text)
```

```python
df.columns
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

```

```python
vectors=cv.fit_transform(df['Clean_Article']).toarray()
vectors
```

```python
# by using cosine Smilarity
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity.shape
```

```python
similarity[0]
```

```python
# Highest Similarity of first 5 Article Sorted
sorted(similarity[0],reverse=True)
```

```python
#sorted by seconde number
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def Recommende(Article):
    Article_idx=df[df['Title']==Article].index[0]
    distances=similarity[Article_idx]
    Article_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in Article_list :
        print(df.iloc[i[0]].Title)
```

```python
Recommende("Best Books to Learn NLP")
```

```python
Recommende('News Classification with Machine Learning')
```
